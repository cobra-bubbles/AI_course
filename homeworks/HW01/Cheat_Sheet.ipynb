## 1. Работа с гитом
### 1.1 Клонировать репозиторий на компьютер
```bash
git clone https://github.com/ORG/REPO.git
```

### 1.2 Локально сохраняет информацию обо всех пушах
```bash
git config --global user.name "Имя Фамилия"
git config --global user.email "your_email@example.com"
```
config - это команда "настроек", где можно менять не только user.name или user.email. --global - значит меняем глобально на всем компьютере, а не только локально на проекте.

### 1.3 Проверить настройки гита
```bash
git config --list
```

**Вывод**
```bash
user.name=NikitaSemin
user.email=ysmalos@list.ru
```

Без настройки информация о пушах будет такой:
```bash
Author: Unknown <unknown@example.com>
```

### 1.4 Локальный статус репозитория
```bash
git status
```

Git покажет:
- какие файлы изменены;
- какие не отслеживаются;
- какие уже подготовлены к коммиту.

### 1.5 Локальные изменения
```bash
git add path/to/file.py
git commit -m "Краткое описание изменений"
```

### 1.6 Пуш в репозиторий
```bash
git push
```

### 1.7 Обновление репозитория локально
```bash
git pull
```

### 1.8 Восстановление документа
```bash
git restore *file_name*
```

Эта команда подтянет все изменения с глобального репозитория в локальный.
## 2 Работа с виртуалкой
### 2.1 Создать виртуальное окружение
```bash
python -m venv .venv
```

### 2.2 Активировать виртуальное окружение
```bash
source .venv/bin/activate
```

### 2.3 Декативировать виртуальное окружение
```bash
deactivate
```
