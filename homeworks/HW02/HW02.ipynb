```python
data.groupby("country").agg(
    median_age=("age", 'median'),
    median_purchases=("purchases", 'median')
)
```

**Также можно разбить возраст на несколько групп и сделать агрегацию по ним**

```python
age_up_limit = data["age"].quantile(0.99)
age_down_limit = data["age"].quantile(0.01)
reasonable_age = data[(data["age"] > age_down_limit) & (data["age"] < age_up_limit)].copy()
cuted_age = pd.cut(reasonable_age["age"], bins=5, precision=0, include_lowest=True)
reasonable_age["age_board"] = cuted_age
reasonable_age.groupby("age_board").agg(count=("age", 'count'))
```

**Можно наблюдать, что возраст 17-26 лет доминирует. Пока мы строили этот series, мы отбросили аномальный возраст (120 лет и 9 лет с покупками и зарплатой). Также можно отметить, что американцы совершают в среднем меньше всего покупок, у китайцев наименьший средний возраст покупок.**

# 2 Визуализация данных в Matplotlib
## 2.1 Гистограмма

```python
data_without_na = data[~data["age"].isna()].copy()

fig, (ax1, ax2) = plt.subplots(
    nrows=1, ncols=2,
    figsize=(8, 4)
)

sns.histplot(data_without_na["age"], ax=ax1)
ax1.set_title("Гистограмма покупок с выбросами")
sns.histplot(reasonable_age["age"], ax=ax2)
ax2.set_title("Гистограмма покупок без выбросов")

plt.tight_layout()
fig.savefig('figures/age_displot.png')
```

**Можно наблюдать красивую и приятную глазу гистограмму без выбросов. Можно сделать вывод, что в 24-29 лет случаются наибольшие покупки.**

```python
fig, (ax1, ax2) = plt.subplots(
    nrows=1, ncols=2,
    figsize=(8, 4)
)

age_data = data[~data["age"].isna()]["age"]
sns.boxplot(
    data=age_data,
    orient="h",
    ax=ax1
)
sns.boxplot(
    data=data["revenue"],
    orient="h",
    ax=ax2
)
plt.suptitle("Ящики с усами")
plt.tight_layout()
fig.savefig('figures/boxplot.png')
```

**Можно наблюдать красивую и приятную глазу гистограмму без выбросов. Можно сделать вывод, что возраст расположен относительно плотно и выбросов немного (в количество одной единицы, 120 лет). С revenue немного иначе, хотя выбросов тоже немного. Разброс больше.**

```python
trace1 = go.Scatter(
    x=data["age"],
    y=data["revenue"],
    name='Revenue based on age'
)
trace2 = go.Scatter(
    x=data["purchases"],
    y=data["revenue"],
    name='Revenue based on age'
)
data = [trace1, trace2]
layout = {'title': 'Scatter'}
fig = go.Figure(data=data, layout=layout)
iplot(fig, show_link=False)
fig.write_html("figures/scatter.html")
```

**Можно наблюдать абсолютную белиберду и сделать вывод, что корреляции revenue от возраста или покупок может и есть, но на данном графике ее наблюдать проблематично.**
