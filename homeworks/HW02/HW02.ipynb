```python
print("Количество ненормальных возрастов: ", data[data["age"] <= 0]["age"].count() + data[data["age"] >= 100]["age"].count())
data[data["age"] >= 100]
```

```python
print("Количество ненормальных покупок: ", data[data["purchases"] < 0]["purchases"].count())
data[data["purchases"] < 0]
```

```python
anomal_user_id = data.groupby("user_id")
for user_is, ser in anomal_user_id:
    if ser["age"].count() > 1:
        print("Ненормальный user_id: ", user_is)
        print(ser)
```

```python
print("Покупки без зарплаты:\n ", data[(data["revenue"] == 0) & (data["purchases"] > 0)])
```

**Были обнаружены некоторые аномалии: отрицательные покупки, слишком большой возраст, покупки без зарплаты, дублирующийся user_id**
