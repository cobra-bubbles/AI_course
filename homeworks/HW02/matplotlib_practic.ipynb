```python
rng = np.arange(50)
rnd = np.random.randint(0, 10, size=(3, rng.size))
yrs = 1950 + rng

fig, ax = plt.subplots(figsize=(5, 3)) #    
ax.stackplot(yrs, rng + rnd, labels=['Eastasia', 'Eurasia', 'Oceania']) #   ,     ,     ()
ax.set_title('Combined debt growth over time') #   
ax.legend(loc='upper left') #   ,   
ax.set_ylabel('Total debt') #   
ax.set_xlim(xmin=yrs[0], xmax=yrs[-1]) #     
fig.tight_layout() #     

plt.show()
```

```python
x = np.random.randint(low=1, high=11, size=50) #
y = x + np.random.randint(1, 5, size=x.size) #
data = np.column_stack((x, y)) #

fig, (ax1, ax2) = plt.subplots( #    
    nrows=1, ncols=2, #      
    figsize=(8, 4) #  
) #

ax1.scatter(x=x, y=y, marker='o', c='r', edgecolor='b') #     
ax1.set_title('Scatter: $x$ versus $y$') #   
ax1.set_xlabel('$x$') #     ,  
ax1.set_ylabel('$y$') #     ,  

ax2.hist( #    - 
    data, bins=np.arange(data.min(), data.max()), #    
    label=('x', 'y') #   
)

ax2.legend(loc=(0.65, 0.8)) #      
ax2.set_title('Frequencies of $x$ and $y$') #   
ax2.yaxis.tick_right() # ,      

plt.show()
```

```python
x = np.diag(np.arange(2, 12))[::-1]
x[np.diag_indices_from(x[::-1])] = np.arange(2, 12)
x2 = np.arange(x.size).reshape(x.shape)
print(x,'\n\n',x2)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 4))
img1=ax1.matshow(x)
ax1.axis('off')
img2 = ax2.matshow(x2, cmap='RdYlGn_r') #
ax2.axis('off')

for i, j in zip(*x.nonzero()): #
    ax1.text(j, i, x[i, j], color='white', ha='center', va='center') #

plt.colorbar(img2, ax=[ax2], shrink=0.79, pad=0.01)
plt.colorbar(img1, ax=[ax1], shrink=0.79, pad=0.01)
fig.suptitle('Heatmaps with `Axes.matshow`', fontsize=16) #

plt.show() #
```

```python
import numpy as np # numpy 
import matplotlib.pyplot as plt #  

import pandas as pd # pandas  

url = 'https://fred.stlouisfed.org/graph/fredgraph.csv?id=VIXCLS' # ,   
vix = pd.read_csv(url,# 
                  index_col=0, #    
                  parse_dates=True, #     
                  na_values='.', #    
                  infer_datetime_format=True, #      ,    
                  #squeeze=True #  Series     
                 ).dropna()#     
vix=vix.squeeze() #  Series     
ma = vix.rolling('90d').mean() #      (90)       
state = pd.cut( #    
                ma, #  
                bins=[-np.inf, 14, 18, 24, np.inf], # 
                labels=range(4) #    (   0  3)
                )

cmap = plt.get_cmap('RdYlGn_r') #  
ma.plot( #  
    color='black',#  
    linewidth=1.5,#  
    marker='',#  ()
    figsize=(8, 4),#  
    label='VIX 90d MA'#  
    )

ax = plt.gca()  #    Axes,    ma.plot()
ax.set_xlabel('') #    ()
ax.set_ylabel('90d moving average: CBOE VIX') #   
ax.set_title('Volatility Regime State') #   
ax.grid(False) #    
ax.set_xlim(xmin=ma.index[0], xmax=ma.index[-1]) #      ()

trans=ax.get_xaxis_transform()#  ,   -  ,   -  .
#
#for i, color in enumerate(cmap([0.2])): #     state (    cmap).
for i, color in enumerate(cmap([0.2, 0.4, 0.6, 0.8])): #     state (    cmap).
    # enumerate      
    ax.fill_between( #      
                    ma.index,#    
                    0,# y    ()
                    1,#     ()
                    where=state == i, #  ,   , ..  ,  state    
                    facecolor=color, #  
                    transform=trans) #  

ax.axhline( #    
    vix.mean(), #     
    linestyle='dashed',  # 
    color='xkcd:dark grey', # 
    alpha=0.6, # 
    label='Full-period mean', #  
    marker='')#   ()

ax.legend(loc='upper left') #   

plt.show() #
```
